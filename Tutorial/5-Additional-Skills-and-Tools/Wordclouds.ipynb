# Creating Word Clouds

### With the [WordCloud](https://amueller.github.io/word_cloud/index.html) package, you can generate massive word clouds images and then choose your favourate! 
### Steps are simple: 
- provide input text
- create a mask 
- choose a plotting scheme
- repeat the same codes multiple times and choose your favourate image

### Here I show how I created the word cloud image on the welcome page of this tutorial:

```python
##############################################################################################
# load packages
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# provide some input text
text = "Python,Xarray,Pandas,NumPy,SciPy,Matplotlib,GeoViews,Dask,netCDF,\
        h5py,Cartopy,Geopandas,SpatialPandas,Plotly,Bokeh,Seaborn,GCPy,Leafmap,Prettymaps,Datashader,HoloViews"

# create a matplotlib figure
fig = plt.figure(figsize=[15,5])

# create a mask to decide the dimensions of the output word clouds image
x, y = np.ogrid[:400,:1200]
mask = (x-200)**2 + (y-600) ** 2 == 400*1200
mask = 255 * mask.astype(int)

# create a word cloud with white background
wc = WordCloud(background_color="white", repeat=True, mask=mask)

# add your text to the word cloud
wc.generate(text)

# display the result
plt.axis("off")
plt.imshow(wc, interpolation="bilinear") 
plt.show()

# Other "interpolation" are also provided by matplotlib,
# but it seems that the interplotation method does not affect the word cloud images.
```

```python
# you can modify the dimensions of output figure and mask to change the shape of your word clouds

fig = plt.figure(figsize=[10,10])

x, y = np.ogrid[:500,:500]
mask = (x-250)**2 + (y-250)**2 > 200**2
mask = 255 * mask.astype(int)
wc = WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
# you can also modify the background color

fig = plt.figure(figsize=[10,10])

x, y = np.ogrid[:500,:500]
mask = (x-250)**2 + (y-250)**2 > 200**2
mask = 255 * mask.astype(int)
wc = WordCloud(background_color="black", repeat=True, mask=mask)
wc.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

### You can also create masked word clouds, image-colored word clouds and a few other types of images. Just refer to examples on [WordCloud](https://amueller.github.io/word_cloud/index.html) homepage.
